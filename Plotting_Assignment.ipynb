```python
# histogram of 27 bin CO2 emissions
df['co2'] = pd.to_numeric(df['co2'], errors='coerce')
df['city'] = pd.to_numeric(df['city'], errors='coerce')
df['comb'] = pd.to_numeric(df['comb'], errors='coerce')

plt.figure(figsize=(10, 6))
plt.hist(df['co2'], bins=27, color='skyblue', edgecolor='black')
plt.title('Carbon Dioxide Emissions')
plt.xlabel('CO2 Emissions')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')

correlation_matrix = numeric_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Heatmap of Correlation ')
plt.show()
```

1.**What story does the visualization tell?**

 The heatmap shows that as CO2 emissions, engine size, and cylinders increase, the ghg_score decreases. It also shows strong relationships between city, highway, and combined, meaning they give similar information. This helps us see which features affect the ghg_score most and which ones might be removed.

2**.What Columns would you delete Drop if the target is the ghg score?**

 To predict ghg_score, we should drop columns like co2, comb, and city because they are directly used to calculate the score, which causes data leakage. We also drop text columns like make and model because they’re hard to use without extra processing. Instead, we keep simple numbers like engine_size and cylinders that help predict emissions naturally.
